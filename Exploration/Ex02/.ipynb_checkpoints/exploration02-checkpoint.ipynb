## 1. Oxford-IIIT Pet 데이터셋을 불러오기 

```python
# Pytorch
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchinfo import summary
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 전처리 정의 (CIFAR-10과 다르게 OxfordIIITPet은 이미지 크기가 다름)
transform = transforms.Compose([
    transforms.Resize((224, 224)),         # 모든 이미지를 224x224로 맞춤
    transforms.ToTensor(),                # 0~1 사이 텐서로 변환
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)) # 정규화
])
```

```python
# 2. 데이터셋 다운로드 및 생성
# 학습 데이터
trainset = torchvision.datasets.OxfordIIITPet(
    root='./data', 
    split='trainval', # CIFAR-10과 다르게 
    download=True, 
    transform=transform
)

# 테스트 데이터
testset = torchvision.datasets.OxfordIIITPet(
    root='./data', 
    split='test', 
    download=True, 
    transform=transform
)
```

```python
# 3. 데이터로더 설정
trainloader = torch.utils.data.DataLoader(trainset, batch_size=32, shuffle=True)
testloader = torch.utils.data.DataLoader(testset, batch_size=32, shuffle=False)

# 데이터셋을 로드하고 기본 정보를 
image, label = trainset[0]

print(f"Image shape: {image.shape}")
print(f"Label: {label}")
print(f"Number of classes: {len(trainset.classes)}")
```

```python
# 데이터의 개수도 확인
train_size = torch.tensor(len(trainset))
test_size = torch.tensor(len(testset))

print(f"Train dataset size: {train_size} (Shape: {train_size.shape})")
print(f"Test dataset size: {test_size} (Shape: {test_size.shape})")
```

```python
# 다운받은 이미지 시각화 하기
def imshow(img):
    img = img / 2 + 0.5
    npimg = img.numpy()
    return np.transpose(npimg, (1, 2, 0))

def show_multiple_images(dataset, n_images=9):
    dataiter = iter(dataset)
    images, labels = next(dataiter)
    fig, axes = plt.subplots(3, 3, figsize=(6, 6))
    axes = axes.flatten()

    for i in range(n_images):
        ax = axes[i]
        img = imshow(images[i])
        ax.imshow(img)
        ax.set_title(f"Label: {trainset.classes[labels[i]]}")
        ax.axis('off')

    plt.tight_layout()
    plt.show()

show_multiple_images(trainloader)
```

## 2. ResNet-34 구현하기

### (1) 블록 구성하기
ResNet 논문에서 아래와 같은 Block 구성 표가 있었다.
![ResNet-34 Block 구성](./ResNet_34.jpg)

```python
# ResNet34 모델 생성하는 클래스
# block_type이 BasicBlock이면 ResNet-34, Bottleneck이면 ResNet-50
class ResNet34(nn.Module):
    def __init__(self, num_blocks_list=[3, 4, 6, 3], channel_list=[64, 128, 256, 512], num_classes=37):
        super(ResNet34, self).__init__()

        # conv1
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False), # bias는 batch normalization을 하기 때문에 필요 없음.
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True) # 덮어쓰기=True
        )
        self.maxpool = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)  

        # conv2 ~ conv5
        # 각 conv 블록들과 shortcut들을 생성
        in_channels = 64
        self.conv_blocks = nn.ModuleList()
        self.conv_shortcuts = nn.ModuleList()

        # conv2_x: 64→64, 크기 유지 (stride=1), 블록 3개
        blocks2, shortcuts2 = self._make_resnet34_block(in_channels=64, out_channels=64, num_blocks=3, stride=1)
        self.conv_blocks.append(blocks2)
        self.conv_shortcuts.append(shortcuts2)
        
        # conv3_x: 64→128, 크기 절반 (stride=2), 블록 4개
        blocks3, shortcuts3 = self._make_resnet34_block(in_channels=64, out_channels=128, num_blocks=4, stride=2)
        self.conv_blocks.append(blocks3)
        self.conv_shortcuts.append(shortcuts3)
        
        # conv4_x: 128→256, 크기 절반 (stride=2), 블록 6개
        blocks4, shortcuts4 = self._make_resnet34_block(in_channels=128, out_channels=256, num_blocks=6, stride=2)
        self.conv_blocks.append(blocks4)
        self.conv_shortcuts.append(shortcuts4)
        
        # conv5_x: 256→512, 크기 절반 (stride=2), 블록 3개
        blocks5, shortcuts5 = self._make_resnet34_block(in_channels=256, out_channels=512, num_blocks=3, stride=2)
        self.conv_blocks.append(blocks5)
        self.conv_shortcuts.append(shortcuts5)

        # 7×7 feature map을 1×1로 줄임
        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))

        # fully connected layer
        self.fc = nn.Linear(512, num_classes)

    # conv blocks 만들기
    def _make_resnet34_block(self, in_channels, out_channels, num_blocks, stride):
        blocks = nn.ModuleList()      # F(x) 경로 (conv 레이어들) 저장
        shortcuts = nn.ModuleList()   # shortcut 경로 (x 경로) 저장

        for i in range(num_blocks):
            # 첫 번째 블록만 크기 축소 + 채널 변경 담당
            if i == 0:
                s = stride
                in_ch = in_channels 
            else:
                s = 1
                in_ch = out_channels

            # conv 2개 거치는 부분 : F(x)
            block = nn.Sequential(
                # 첫 번째 conv: 크기 축소가 필요하면 여기서 stride로 처리
                nn.Conv2d(in_ch, out_channels, kernel_size=3, stride=s, padding=1, bias=False),
                nn.BatchNorm2d(out_channels),
                nn.ReLU(inplace=True),
    
                # 두 번째 conv: 항상 크기 유지 (stride=1)
                nn.Conv2d(out_channels, out_channels, kernel_size=3, stride=1, padding=1, bias=False),
                nn.BatchNorm2d(out_channels),
                # 여기서 ReLU 안 함 → forward에서 identity 더한 후에 ReLU 적용
            )
            blocks.append(block)
    
            # shortcut 부분 : x
            if s != 1 or in_ch != out_channels:
                # 크기나 채널이 달라서 x를 그대로 더할 수 없는 경우 (conv2 -> conv3로 넘어갈 때)
                # 1x1 conv로 채널 수 맞추고, stride로 크기도 맞춤
                shortcut = nn.Sequential(
                    nn.Conv2d(in_ch, out_channels, kernel_size=1, stride=s, bias=False),
                    nn.BatchNorm2d(out_channels)
                )
            else:
                # 크기와 채널이 같아서 x를 그대로 더할 수 있는 경우 (conv2 안에서 다음 블럭으로 넘어갈 때)
                # Identity: 아무것도 안 하고 그대로 통과
                shortcut = nn.Identity()
            shortcuts.append(shortcut)
        return blocks, shortcuts

    def forward(self, x):
        # conv1: 7x7 conv + max pooling
        x = self.conv1(x)       # 224x224 → 112x112
        x = self.maxpool(x)     # 112x112 → 56x56
    
        # conv2_x ~ conv5_x: 각 블록마다 residual 연산
        for blocks, shortcuts in zip(self.conv_blocks, self.conv_shortcuts):
            for block, shortcut in zip(blocks, shortcuts):
                identity = shortcut(x)           # shortcut 경로: x (또는 1x1 conv로 변환된 x)
                x = F.relu(block(x) + identity)  # F(x) + x 한 후 ReLU
    
        # 분류기: average pool + fc (논문 Table 1 하단)
        x = self.avgpool(x)         # 7x7 → 1x1
        x = torch.flatten(x, 1)    # (배치, 512, 1, 1) → (배치, 512)로 펴줌
        x = self.fc(x)              # 512 → 37 (클래스 점수)
        return x
```

```python
resnet_34 = ResNet34(num_classes=37)
summary(resnet_34, input_size=(1, 3, 224, 224))
```

### (2) 위에서 만든 모델로 학습

```python
import time

current_time = time.time()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 모델 생성 (pretrained 아님, 처음부터 학습)
resnet34 = ResNet34(num_classes=37)
resnet34.to(device)

# 논문 Section 3.4: SGD, momentum 0.9, weight decay 0.0001
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(resnet34.parameters(), lr=0.01, momentum=0.9, weight_decay=1e-4)

# 논문: 성능 정체 시 학습률을 10으로 나눔
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5)

resnet34_train_losses = []
resnet34_val_accuracy = []
EPOCH = 30

for epoch in range(EPOCH):
    resnet34.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for i, (inputs, labels) in enumerate(trainloader, 0):
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = resnet34(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        if i % 100 == 99:
            print(f"[{epoch + 1}, {i + 1:5d}] loss: {running_loss / (i + 1):.3f}")

    train_loss = running_loss / len(trainloader)
    train_acc = 100 * correct / total
    resnet34_train_losses.append(train_loss)
    print(f"Epoch {epoch + 1}: Train Accuracy: {train_acc:.2f}%")

    # 학습률 스케줄러 업데이트 (train_loss 기준으로 정체 판단)
    scheduler.step(train_loss)

    resnet34.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in testloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = resnet34(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_acc = 100 * correct / total
    resnet34_val_accuracy.append(val_acc)
    print(f"Epoch {epoch + 1}: Validation Accuracy: {val_acc:.2f}%")

print("Finished Training")
print(time.time() - current_time)
```

## 3. ResNet-50 구현하기

### (1) 블록 구성하기
ResNet 논문에서 아래와 같은 Block 구성 표가 있었다.
![ResNet-50 Block 구성](./ResNet_50.jpg)

```python
class ResNet50(nn.Module):
    def __init__(self, num_blocks_list=[3, 4, 6, 3], channel_list=[64, 128, 256, 512], num_classes=37):
        super(ResNet50, self).__init__()
        
        # expansion: bottleneck의 마지막 1x1 conv가 채널을 4배로 늘림
        # 예: channel 64 → 마지막 출력은 64 * 4 = 256
        self.expansion = 4

        # conv1: ResNet-34와 동일
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True)
        )
        self.maxpool = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)

        # conv2 ~ conv5
        self.conv_blocks = nn.ModuleList()
        self.conv_shortcuts = nn.ModuleList()

        # conv2_x: 64→256, 크기 유지 (stride=1), 블록 3개
        # in_channels=64 (conv1 출력), mid_channels=64, out_channels=64*4=256
        blocks2, shortcuts2 = self._make_resnet50_block(in_channels=64, mid_channels=64, num_blocks=3, stride=1)
        self.conv_blocks.append(blocks2)
        self.conv_shortcuts.append(shortcuts2)

        # conv3_x: 256→512, 크기 절반 (stride=2), 블록 4개
        blocks3, shortcuts3 = self._make_resnet50_block(in_channels=256, mid_channels=128, num_blocks=4, stride=2)
        self.conv_blocks.append(blocks3)
        self.conv_shortcuts.append(shortcuts3)

        # conv4_x: 512→1024, 크기 절반 (stride=2), 블록 6개
        blocks4, shortcuts4 = self._make_resnet50_block(in_channels=512, mid_channels=256, num_blocks=6, stride=2)
        self.conv_blocks.append(blocks4)
        self.conv_shortcuts.append(shortcuts4)

        # conv5_x: 1024→2048, 크기 절반 (stride=2), 블록 3개
        blocks5, shortcuts5 = self._make_resnet50_block(in_channels=1024, mid_channels=512, num_blocks=3, stride=2)
        self.conv_blocks.append(blocks5)
        self.conv_shortcuts.append(shortcuts5)

        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(512 * self.expansion, num_classes)  # 2048 → 37

    def _make_resnet50_block(self, in_channels, mid_channels, num_blocks, stride):
        """
        ResNet-34의 _make_resnet34_block에 대응
        차이점: conv 2개(3x3, 3x3) 대신 conv 3개(1x1, 3x3, 1x1) 사용
        """
        blocks = nn.ModuleList()
        shortcuts = nn.ModuleList()
        out_channels = mid_channels * self.expansion  # 최종 출력 채널 (예: 64*4=256)

        for i in range(num_blocks):
            if i == 0:
                s = stride
                in_ch = in_channels       # 이전 stage 출력 (예: conv2_x 첫 블록은 64)
            else:
                s = 1
                in_ch = out_channels      # 이전 블록 출력 (예: 256)

            # ============================================
            # Bottleneck 블록: 1x1 → 3x3 → 1x1 (논문 Figure 5 오른쪽)
            # 1x1: 채널 축소 (256 → 64)
            # 3x3: 특징 추출 (64 → 64)
            # 1x1: 채널 확장 (64 → 256)
            # ============================================
            block = nn.Sequential(
                # 첫 번째 1x1 conv: 채널 축소 (예: 256 → 64)
                nn.Conv2d(in_ch, mid_channels, kernel_size=1, stride=1, bias=False),
                nn.BatchNorm2d(mid_channels),
                nn.ReLU(inplace=True),

                # 두 번째 3x3 conv: 특징 추출 (크기 축소는 여기서)
                nn.Conv2d(mid_channels, mid_channels, kernel_size=3, stride=s, padding=1, bias=False),
                nn.BatchNorm2d(mid_channels),
                nn.ReLU(inplace=True),

                # 세 번째 1x1 conv: 채널 확장 (예: 64 → 256)
                nn.Conv2d(mid_channels, out_channels, kernel_size=1, stride=1, bias=False),
                nn.BatchNorm2d(out_channels),
                # ReLU 안 함 → forward에서 identity 더한 후에 적용
            )
            blocks.append(block)

            # shortcut: 차원 맞추기
            if s != 1 or in_ch != out_channels:
                shortcut = nn.Sequential(
                    nn.Conv2d(in_ch, out_channels, kernel_size=1, stride=s, bias=False),
                    nn.BatchNorm2d(out_channels)
                )
            else:
                shortcut = nn.Identity()
            shortcuts.append(shortcut)

        return blocks, shortcuts

    def forward(self, x):
        x = self.conv1(x)       # 224 → 112
        x = self.maxpool(x)     # 112 → 56

        for blocks, shortcuts in zip(self.conv_blocks, self.conv_shortcuts):
            for block, shortcut in zip(blocks, shortcuts):
                identity = shortcut(x)
                x = F.relu(block(x) + identity)

        x = self.avgpool(x)         # 7 → 1
        x = torch.flatten(x, 1)    # (배치, 2048, 1, 1) → (배치, 2048)
        x = self.fc(x)              # 2048 → 37
        return x


# 모델 생성 및 확인
model50 = ResNet50(num_classes=37)
dummy = torch.randn(1, 3, 224, 224)
print(model50(dummy).shape)  # torch.Size([1, 37])
```

```python
resnet_50 = ResNet50(num_classes=37)
summary(resnet_50, input_size=(1, 3, 224, 224))
```

### (2) 위에서 만든 모델로 학습

```python
import time
current_time = time.time()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# 모델 생성 (pretrained 아님, 처음부터 학습)
resnet50 = ResNet50(num_classes=37)
resnet50.to(device)
# 논문 Section 3.4: SGD, momentum 0.9, weight decay 0.0001
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(resnet50.parameters(), lr=0.01, momentum=0.9, weight_decay=1e-4)
# 논문: 성능 정체 시 학습률을 10으로 나눔
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5)
resnet50_train_losses = []
resnet50_val_accuracy = []
EPOCH = 30
for epoch in range(EPOCH):
    resnet50.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for i, (inputs, labels) in enumerate(trainloader, 0):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = resnet50(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        if i % 100 == 99:
            print(f"[{epoch + 1}, {i + 1:5d}] loss: {running_loss / (i + 1):.3f}")
    train_loss = running_loss / len(trainloader)
    train_acc = 100 * correct / total
    resnet50_train_losses.append(train_loss)
    print(f"Epoch {epoch + 1}: Train Accuracy: {train_acc:.2f}%")
    # 학습률 스케줄러 업데이트 (train_loss 기준으로 정체 판단)
    scheduler.step(train_loss)
    resnet50.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in testloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = resnet50(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    val_acc = 100 * correct / total
    resnet50_val_accuracy.append(val_acc)
    print(f"Epoch {epoch + 1}: Validation Accuracy: {val_acc:.2f}%")
print("Finished Training")
print(time.time() - current_time)
```

## Plain 34 layer, 50 layer 구현하기

```python
# ============================================
# Plain-34: ResNet-34에서 skip connection 제거
# ============================================
class Plain34(nn.Module):
    def __init__(self, num_blocks_list=[3, 4, 6, 3], channel_list=[64, 128, 256, 512], num_classes=37):
        super(Plain34, self).__init__()

        # conv1: ResNet-34와 동일
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True)
        )
        self.maxpool = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)

        # conv2 ~ conv5: shortcut 없이 블록만 생성
        self.conv_blocks = nn.ModuleList()

        # conv2_x: 64→64, 크기 유지, 블록 3개
        self.conv_blocks.append(self._make_plain34_block(in_channels=64, out_channels=64, num_blocks=3, stride=1))
        # conv3_x: 64→128, 크기 절반, 블록 4개
        self.conv_blocks.append(self._make_plain34_block(in_channels=64, out_channels=128, num_blocks=4, stride=2))
        # conv4_x: 128→256, 크기 절반, 블록 6개
        self.conv_blocks.append(self._make_plain34_block(in_channels=128, out_channels=256, num_blocks=6, stride=2))
        # conv5_x: 256→512, 크기 절반, 블록 3개
        self.conv_blocks.append(self._make_plain34_block(in_channels=256, out_channels=512, num_blocks=3, stride=2))

        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(512, num_classes)

    def _make_plain34_block(self, in_channels, out_channels, num_blocks, stride):
        """ResNet-34의 _make_resnet34_block에서 shortcut 제거한 버전"""
        blocks = nn.ModuleList()

        for i in range(num_blocks):
            if i == 0:
                s = stride
                in_ch = in_channels
            else:
                s = 1
                in_ch = out_channels

            # conv 2개 (3x3 → 3x3) + ReLU까지 포함
            # ResNet과 차이: skip connection이 없으니 마지막에 바로 ReLU
            block = nn.Sequential(
                nn.Conv2d(in_ch, out_channels, kernel_size=3, stride=s, padding=1, bias=False),
                nn.BatchNorm2d(out_channels),
                nn.ReLU(inplace=True),

                nn.Conv2d(out_channels, out_channels, kernel_size=3, stride=1, padding=1, bias=False),
                nn.BatchNorm2d(out_channels),
                nn.ReLU(inplace=True),  # ← ResNet은 여기서 identity 더한 후 ReLU, Plain은 바로 ReLU
            )
            blocks.append(block)

        return blocks

    def forward(self, x):
        x = self.conv1(x)       # 224 → 112
        x = self.maxpool(x)     # 112 → 56

        # skip connection 없이 그냥 통과
        for stage_blocks in self.conv_blocks:
            for block in stage_blocks:
                x = block(x)    # ← identity 더하기 없음!

        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.fc(x)
        return x


# ============================================
# Plain-50: ResNet-50에서 skip connection 제거
# ============================================
class Plain50(nn.Module):
    def __init__(self, num_blocks_list=[3, 4, 6, 3], channel_list=[64, 128, 256, 512], num_classes=37):
        super(Plain50, self).__init__()

        self.expansion = 4

        # conv1: ResNet-50와 동일
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True)
        )
        self.maxpool = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)

        # conv2 ~ conv5: shortcut 없이 블록만 생성
        self.conv_blocks = nn.ModuleList()

        # conv2_x: 64→256, 크기 유지, 블록 3개
        self.conv_blocks.append(self._make_plain50_block(in_channels=64, mid_channels=64, num_blocks=3, stride=1))
        # conv3_x: 256→512, 크기 절반, 블록 4개
        self.conv_blocks.append(self._make_plain50_block(in_channels=256, mid_channels=128, num_blocks=4, stride=2))
        # conv4_x: 512→1024, 크기 절반, 블록 6개
        self.conv_blocks.append(self._make_plain50_block(in_channels=512, mid_channels=256, num_blocks=6, stride=2))
        # conv5_x: 1024→2048, 크기 절반, 블록 3개
        self.conv_blocks.append(self._make_plain50_block(in_channels=1024, mid_channels=512, num_blocks=3, stride=2))

        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(512 * self.expansion, num_classes)  # 2048 → 37

    def _make_plain50_block(self, in_channels, mid_channels, num_blocks, stride):
        """ResNet-50의 _make_resnet50_block에서 shortcut 제거한 버전"""
        blocks = nn.ModuleList()
        out_channels = mid_channels * self.expansion  # 예: 64*4=256

        for i in range(num_blocks):
            if i == 0:
                s = stride
                in_ch = in_channels
            else:
                s = 1
                in_ch = out_channels

            # Bottleneck: 1x1 → 3x3 → 1x1 + 마지막에 바로 ReLU
            block = nn.Sequential(
                # 1x1: 채널 축소 (예: 256 → 64)
                nn.Conv2d(in_ch, mid_channels, kernel_size=1, stride=1, bias=False),
                nn.BatchNorm2d(mid_channels),
                nn.ReLU(inplace=True),

                # 3x3: 특징 추출 (크기 축소는 여기서)
                nn.Conv2d(mid_channels, mid_channels, kernel_size=3, stride=s, padding=1, bias=False),
                nn.BatchNorm2d(mid_channels),
                nn.ReLU(inplace=True),

                # 1x1: 채널 확장 (예: 64 → 256)
                nn.Conv2d(mid_channels, out_channels, kernel_size=1, stride=1, bias=False),
                nn.BatchNorm2d(out_channels),
                nn.ReLU(inplace=True),  # ← ResNet은 identity 더한 후 ReLU, Plain은 바로 ReLU
            )
            blocks.append(block)

        return blocks

    def forward(self, x):
        x = self.conv1(x)       # 224 → 112
        x = self.maxpool(x)     # 112 → 56

        # skip connection 없이 그냥 통과
        for stage_blocks in self.conv_blocks:
            for block in stage_blocks:
                x = block(x)    # ← identity 더하기 없음!

        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.fc(x)
        return x


# 확인
plain34 = Plain34(num_classes=37)
plain50 = Plain50(num_classes=37)
dummy = torch.randn(1, 3, 224, 224)
print(plain34(dummy).shape)  # torch.Size([1, 37])
print(plain50(dummy).shape)  # torch.Size([1, 37])
```

```python
import time

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
criterion = nn.CrossEntropyLoss()
EPOCH = 30

# ============================================
# Plain-34 학습
# ============================================
print("=" * 50)
print("Plain-34 학습 시작")
print("=" * 50)

current_time = time.time()

plain34 = Plain34(num_classes=37)
plain34.to(device)

optimizer = optim.SGD(plain34.parameters(), lr=0.01, momentum=0.9, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5)

plain34_train_losses = []
plain34_val_accuracy = []

for epoch in range(EPOCH):
    plain34.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for i, (inputs, labels) in enumerate(trainloader, 0):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = plain34(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        if i % 100 == 99:
            print(f"[{epoch + 1}, {i + 1:5d}] loss: {running_loss / (i + 1):.3f}")
    train_loss = running_loss / len(trainloader)
    train_acc = 100 * correct / total
    plain34_train_losses.append(train_loss)
    print(f"Epoch {epoch + 1}: Train Accuracy: {train_acc:.2f}%")
    scheduler.step(train_loss)
    plain34.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in testloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = plain34(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    val_acc = 100 * correct / total
    plain34_val_accuracy.append(val_acc)
    print(f"Epoch {epoch + 1}: Validation Accuracy: {val_acc:.2f}%")

print(f"Plain-34 완료! 소요 시간: {time.time() - current_time:.1f}초")

# ============================================
# Plain-50 학습
# ============================================
print("\n" + "=" * 50)
print("Plain-50 학습 시작")
print("=" * 50)

current_time = time.time()

plain50 = Plain50(num_classes=37)
plain50.to(device)

optimizer = optim.SGD(plain50.parameters(), lr=0.01, momentum=0.9, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=5)

plain50_train_losses = []
plain50_val_accuracy = []

for epoch in range(EPOCH):
    plain50.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for i, (inputs, labels) in enumerate(trainloader, 0):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = plain50(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        if i % 100 == 99:
            print(f"[{epoch + 1}, {i + 1:5d}] loss: {running_loss / (i + 1):.3f}")
    train_loss = running_loss / len(trainloader)
    train_acc = 100 * correct / total
    plain50_train_losses.append(train_loss)
    print(f"Epoch {epoch + 1}: Train Accuracy: {train_acc:.2f}%")
    scheduler.step(train_loss)
    plain50.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in testloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = plain50(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    val_acc = 100 * correct / total
    plain50_val_accuracy.append(val_acc)
    print(f"Epoch {epoch + 1}: Validation Accuracy: {val_acc:.2f}%")

print(f"Plain-50 완료! 소요 시간: {time.time() - current_time:.1f}초")

# ============================================
# 전체 완료 요약
# ============================================
print("\n" + "=" * 50)
print("전체 학습 완료! 최종 Validation Accuracy 요약:")
print(f"  Plain-34:  {plain34_val_accuracy[-1]:.2f}%")
print(f"  Plain-50:  {plain50_val_accuracy[-1]:.2f}%")
print("=" * 50)
```

## 결과 분석 및 회고
1. Skip connection의 효과는 명확했다. Plain 대비 ResNet이 34 layers에서 약 21%, 50 layers에서 약 12% 높은 정확도를 보였다. Plain 모델은 레이어를 깊게 쌓을수록 gradient가 제대로 전달되지 않아 학습이 거의 이루어지지 않았고, ResNet은 skip connection 덕분에 gradient 흐름이 유지되어 실제로 학습이 진행되었다.
2. ResNet-50이 ResNet-34보다 낮은 이유를 고민해봤다. 논문에서는 깊을수록 성능이 좋아지지만, 이번 실험에서는 반대 결과가 나왔다. 원인은 OxfordIIITPet 데이터셋이 약 3,680장으로 매우 적기 때문인것 같다. weight가 많아서 더 과적합이 일어나는 것 같다.
3. 더 데이터 셋의 개수가 많은 ImageNet으로 실험을 하거나 이런 데이터가 작을 때 사용하는 데이터 증강이나, regularization 기법을 적용해보면 좋을것 같다는 생각이 들었다.
